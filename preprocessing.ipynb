```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import os
```

```python
import pandas as pd

# Load the Excel file
file_path = "IND.xlsx"

# Read both sheets
sheet1 = pd.read_excel(file_path, sheet_name="Subnational 1 tree cover loss")
sheet2 = pd.read_excel(file_path, sheet_name="Subnational 2 tree cover loss")

# Save as CSV files
sheet1.to_csv("subnational_1_tree_cover_loss.csv", index=False)
sheet2.to_csv("subnational_2_tree_cover_loss.csv", index=False)

print(" Sheets extracted and saved as CSV:")
print("- subnational_1_tree_cover_loss.csv")
print("- subnational_2_tree_cover_loss.csv")

# Quick sanity check: show first few rows
print("\nðŸ”¹ Subnational 1 Preview:")
print(sheet1.head())

print("\nðŸ”¹ Subnational 2 Preview:")
print(sheet2.head())
```

```python
# EDA
print(f"Sheet1 shape: {sheet1.shape}")
print(f"Sheet2 shape: {sheet2.shape}")

#  Standardize Column Names
sheet1.columns = sheet1.columns.str.strip().str.lower().str.replace(" ", "_")
sheet2.columns = sheet2.columns.str.strip().str.lower().str.replace(" ", "_")

print("\nðŸ“‹ Standardized Columns:")
print("Sheet1 Columns:", sheet1.columns.tolist())
print("Sheet2 Columns:", sheet2.columns.tolist())

#  Basic Structure and Info
print("\n--- Sheet1 Info ---")
sheet1.info()
print("\n--- Sheet2 Info ---")
sheet2.info()
```

```python
# missing values
print("\n--- Missing Values in Sheet1 ---")
print(sheet1.isnull().sum())    
print("\n--- Missing Values in Sheet2 ---")
print(sheet2.isnull().sum())

# Handle missing values (example: fill with 0 and text with 'Unknown')
for df in [sheet1, sheet2]:
    for col in df.columns:
        if df[col].dtype in ['float64', 'int64']:
            df[col].fillna(0, inplace=True)
        else:
            df[col].fillna('Unknown', inplace=True)
print("\n Missing values handled.")
```

```python
# Combine both sheets
combined = pd.concat([sheet1, sheet2], ignore_index=True)
print("\nCombined dataset created!")
print("Shape:", combined.shape)

# Summary of numeric data
print("\nSummary statistics (numeric columns):")
print(combined.describe())
```

```python
# Separate numeric and categorical columns
numeric_cols = combined.select_dtypes(include=[np.number]).columns
cat_cols = combined.select_dtypes(exclude=[np.number]).columns

print("\nNumeric columns:", list(numeric_cols))
print("Categorical columns:", list(cat_cols))
```

```python
# Apply KNN Imputation for numeric columns
if not numeric_cols.empty:
    imputer = KNNImputer(n_neighbors=3)
    combined[numeric_cols] = imputer.fit_transform(combined[numeric_cols])
    print("\n KNN imputation applied on numeric columns.")
else:
    print("\n No numeric columns found for KNN imputation.")

```

```python
# Handle missing categorical values
for col in cat_cols:
    combined[col].fillna("Unknown", inplace=True)

print("\n Missing values handled for categorical columns.")

# 8 Apply Label Encoding on categorical columns
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    combined[col] = le.fit_transform(combined[col].astype(str))
    label_encoders[col] = le

print("\n Label encoding done for categorical columns.")
```

```python
#  Summary of text (categorical) data
print("\nTop values in text columns:")
for col in combined.select_dtypes(include="object").columns:
    print(f"\nColumn: {col}")
    print(combined[col].value_counts().head(5))

#  Correlation between numeric columns
numeric_data = combined.select_dtypes(include=[np.number])
if not numeric_data.empty:
    print("\nCorrelation between numeric columns:")
    print(numeric_data.corr())
else:
    print("\nNo numeric columns found for correlation check.")

#  Check min and max values to understand ranges
print("\nMin and max values for numeric columns:")
for col in numeric_data.columns:
    print(f"{col}: Min={numeric_data[col].min()}  |  Max={numeric_data[col].max()}")
```
